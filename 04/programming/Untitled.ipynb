```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [15, 10]
```

```python
n = 100
b0 = 1
b = 3
X = np.random.normal(0, 0.4, size=n)
y = b0 + b * X + np.random.normal(0, 0.5, size=n)
```

```python
splits = [
    (-1 <= X) & (X < -1/3),
    (-1/3 <= X) & (X < 1/3),
    (1/3 <= X) & (X < 1)
]
models = []
for s in splits:
    model = LinearRegression()
    model.fit(X[s].reshape(-1,1),y[s])
    models.append(model)
```

```python
plt.figure()
for i, s in enumerate(splits):
    plt.scatter(X[s], y[s], label=f"S{i+1}")
    plt.plot(X[s], models[i].predict(X[s].reshape(-1,1)))
plt.axvline(x = -1/3, c = "black", ls = "--")
plt.axvline(x = 1/3, c = "black", ls = "--")
plt.xlim(-1,1)
plt.legend()
plt.show()
```
